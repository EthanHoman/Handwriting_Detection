<h1 style="color:darkgoldenrod">AI Problem</h1>

<ul>
<li style="color:maroon"><h4><bf>What is the problem on which you are proposing to work and what are its characteristics?</bf></h4></li>
    <ul>
        <li>Our problem is developing an image recognition model that is capable of correctly identifing handwritten numbers or letters. </li>
        <li>This problem involves analyzing large datasets of images corrosponding to each digit or letter used to train our model.</li>
        <li>The problem's characteristics include multi-class classification (47 classes in the EMNIST), dealing with grayscale images of size 28x28 pixels, and handling potential variations in handwriting styles and quality.</li>
        <li>Scope: We'll aim to develop a machine learning model, likely a convolutional neural network (CNN), to perform this classification task.</li>
        <li>Boundaries: We'll use only the EMNIST dataset for training and testing, not incorporating other datasets or real-time handwriting input.</li>
        <li>Constraints: Our constraints include completing the project within our course timeline, using our available computational resources, while developing a model that achieves at least 90% accuracy on the EMNIST test set and is small enough to run on our standard computers.</li>
    </ul>
<li style="color:maroon"><h4><bf>Why do you think AI might be a good choice for a solution to this problem?</bf></h4></li>
    <ul>
        <li>AI, especially machine learning techniques like neural networks, has been proven to perform well on tasks involving pattern recognition, such as identifying features in images.</li>
        <li>AI has the capability of identifying patterns in handwriting, which enables it to decipher large datasets of images. This capability enhances the efficiency of solutions compared to previous techniques.</li>
    </ul>
<li style="color:maroon"><h4><bf>Why did you choose this problem?</bf></h4></li>
    <ul>
        <li>The problem’s relevance to real world applications, such as handwriting recognition and digitization, makes this project both practical and impactful.</li>
        <li>This can be used for automating data entry, digitalizing historical documents, or grading handwritten assignments.</li>
    </ul>

<h2 style="color: teal">From Journal/Conference Papers</h2>

<p>Journal and conference papers report on research to solve problems.  They do not necessarily have tutorial instructions, but they do report on many ideas that were used in the past, the ideas used by the authors to solve the problem in the paper, and ideas for future research.  They can help you narrow down quickly promising ideas to use to solve a problem.  Examples of ideas are using A-star search and developing a modified version of A-star search to reduce the state space by sampling the actions.  Normal data processing, such as cleaning data, or normal steps to solve a problem would not be the ideas for which you are searching as you read the paper.</p>
<p>Another thing that papers have are results that you can use to compare to your solution so you can see if your solution approach has merit.  If your solution does not do as well as solutions reported in the papers, use the papers to find ideas to improve the solution that you have.</p>
<p>You want to look for papers listed in the Scopus database available through the <a href="https://ttu-primo.hosted.exlibrisgroup.com/primo-explore/dbsearch?vid=01TTU">TTU Library's Website</a>, and enter "Scopus" into the search box.  Papers listed in Scopus will be in venues recognized by scholars and the papers you use should have at least 5 citations.  In particular, when reading the papers, look for relationships among the solution ideas, how the authors evaluated their solutions, and what promising ideas the authors think could improve their work.</p>
