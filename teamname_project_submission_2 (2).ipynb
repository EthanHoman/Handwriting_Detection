```python
#code cells for running and evaluating the solution

import torch
import torch.nn as nn
import pandas as pd
from torch.utils.data import DataLoader, Dataset
from torchvision.transforms import ToTensor

#creation of custom data set to load my csv files 
class CustomMNISTDataset(Dataset):
    def __init__(self, name, transform=ToTensor(), label_name="label", is_labeled=True):
        self.data = pd.read_csv(name)  
        self.transform = transform
        self.label_name = label_name
        self.is_labeled = is_labeled

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        #extracting the pixel values
        try: 
            pixel_data = self.data.iloc[idx].drop(self.label_name, errors='ignore').values.astype('float32')
            scaled_pixel = (pixel_data - pixel_data.min()) / (pixel_data.max() - pixel_data.min())
            image = scaled_pixel.reshape(28, 28)
            if self.transform:
                image = self.transform(image)

            if self.is_labeled :
                label = int(self.data.iloc[idx][self.label_name])
                return image, label
            if self.is_labeled and not self.data.iloc[1]:
                print("Code can't be accessed")
            else:
                return image
            
            
        except Exception as e:
            print("Code could not be retrieved: Error")
            return 0


# downloading data-sets:
train_dataset = CustomMNISTDataset(name="train.csv", is_labeled=True)
if (not train_dataset):
    print("Error cant find training data set")
test_dataset = CustomMNISTDataset(name="test.csv", is_labeled=False)
if (not test_dataset):
    print("Error cant find test data set")
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

#simple CNN: 
class Model(nn.Module):
    def __init__(self):
        try:
            super(Model, self).__init__()
            self.f1_layer = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, stride=1, padding=1)
            self.f2_layer = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1, stride=1)
            self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
            self.flatten = nn.Flatten()
            self.fc = nn.Linear(64 * 7 * 7, 10)  
        except:
            print("cant create convolusion model")

    def forward(self, x): 
        x = self.pool(torch.relu(self.f1_layer(x)))
        x = self.pool(torch.relu(self.f2_layer(x)))
        x = self.flatten(x)
        x = self.fc(x)
        return x

# setting the Model to execute
model = Model()
if not model:
    print("Model cant be created.")
model.eval()
for images in test_loader:
        if not test_loader:
            print("Images cant be found in data set")
        outputs = model(images)
        _, predictions = torch.max(outputs, 1)  
        print("Predicted labels:", predictions)
        break  
```
